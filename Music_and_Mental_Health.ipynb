```python
teensAnxiety = 0
teensDepression = 0
teensInsomnia = 0
teensOCD = 0

twentiesAnxiety = 0
twentiesDepression = 0
twentiesInsomnia = 0
twentiesOCD = 0

thirtiesAnxiety = 0
thirtiesDepression = 0
thirtiesInsomnia = 0
thirtiesOCD = 0

fourtiesAnxiety = 0
fourtiesDepression = 0
fourtiesInsomnia = 0
fourtiesOCD = 0

fiftiesAnxiety = 0
fiftiesDepression = 0
fiftiesInsomnia = 0
fiftiesOCD = 0

sixtiesAnxiety = 0
sixtiesDepression = 0
sixtiesInsomnia = 0
sixtiesOCD = 0

seventiesAnxiety = 0
seventiesDepression = 0
seventiesInsomnia = 0
seventiesOCD = 0

for i in range(len(data)):
    if data.loc[i, "Age"] < 20:
        if data.loc[i, "Anxiety"] == True:
            teensAnxiety += 1
        if data.loc[i, "Depression"] == True:
            teensDepression += 1
        if data.loc[i, "Insomnia"] == True:
            teensInsomnia += 1
        if data.loc[i, "OCD"] == True:
            teensOCD += 1
    elif data.loc[i, "Age"] < 30:
        if data.loc[i, "Anxiety"] == True:
            twentiesAnxiety += 1
        if data.loc[i, "Depression"] == True:
            twentiesDepression += 1
        if data.loc[i, "Insomnia"] == True:
            twentiesInsomnia += 1
        if data.loc[i, "OCD"] == True:
            twentiesOCD += 1
    elif data.loc[i, "Age"] < 40:
        if data.loc[i, "Anxiety"] == True:
            thirtiesAnxiety += 1
        if data.loc[i, "Depression"] == True:
            thirtiesDepression += 1
        if data.loc[i, "Insomnia"] == True:
            thirtiesInsomnia += 1
        if data.loc[i, "OCD"] == True:
            thirtiesOCD += 1
    elif data.loc[i, "Age"] < 50:
        if data.loc[i, "Anxiety"] == True:
            fourtiesAnxiety += 1
        if data.loc[i, "Depression"] == True:
            fourtiesDepression += 1
        if data.loc[i, "Insomnia"] == True:
            fourtiesInsomnia += 1
        if data.loc[i, "OCD"] == True:
            fourtiesOCD += 1
    elif data.loc[i, "Age"] < 60:
        if data.loc[i, "Anxiety"] == True:
            fiftiesAnxiety += 1
        if data.loc[i, "Depression"] == True:
            fiftiesDepression += 1
        if data.loc[i, "Insomnia"] == True:
            fiftiesInsomnia += 1
        if data.loc[i, "OCD"] == True:
            fiftiesOCD += 1
    elif data.loc[i, "Age"] < 70:
        if data.loc[i, "Anxiety"] == True:
            sixtiesAnxiety += 1
        if data.loc[i, "Depression"] == True:
            sixtiesDepression += 1
        if data.loc[i, "Insomnia"] == True:
            sixtiesInsomnia += 1
        if data.loc[i, "OCD"] == True:
            sixtiesOCD += 1
    else:
        if data.loc[i, "Anxiety"] == True:
            seventiesAnxiety += 1
        if data.loc[i, "Depression"] == True:
            seventiesDepression += 1
        if data.loc[i, "Insomnia"] == True:
            seventiesInsomnia += 1
        if data.loc[i, "OCD"] == True:
            seventiesOCD += 1
            
totalAnxiety = teensAnxiety + twentiesAnxiety + thirtiesAnxiety + fourtiesAnxiety + fiftiesAnxiety + sixtiesAnxiety + seventiesAnxiety
totalDepression = teensDepression + twentiesDepression + thirtiesDepression + fourtiesDepression + fiftiesDepression + sixtiesDepression + seventiesDepression
totalInsomnia = teensInsomnia + twentiesInsomnia + thirtiesInsomnia + fourtiesInsomnia + fiftiesInsomnia + sixtiesInsomnia + seventiesInsomnia
totalOCD = teensOCD + twentiesOCD + thirtiesOCD + fourtiesOCD + fiftiesOCD + sixtiesOCD + seventiesOCD
            
ages = ["10-19", "20-29", "30-39", "40-49", "50-59", "60-69", "70-80"]
results = {
    "Anxiety" : (teensAnxiety, twentiesAnxiety, thirtiesAnxiety, fourtiesAnxiety, fiftiesAnxiety, sixtiesAnxiety, seventiesAnxiety),
    "Depression" : (teensDepression, twentiesDepression, thirtiesDepression, fourtiesDepression, fiftiesDepression, sixtiesDepression, 
                    seventiesDepression),
    "Insomnia" : (teensInsomnia, twentiesInsomnia, thirtiesInsomnia, fourtiesInsomnia, fiftiesInsomnia, sixtiesInsomnia, seventiesInsomnia),
    "OCD" : (teensOCD, twentiesOCD, thirtiesOCD, fourtiesOCD, fiftiesOCD, sixtiesOCD, seventiesOCD)
}

x = np.arange(len(ages))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in results.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=7)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Diagnosis Count')
ax.set_title('Diagnosis by age range')
ax.set_xticks(x + width, ages)
ax.legend()
ax.set_ylim(0, 250)

plt.show()
```
